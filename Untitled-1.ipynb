```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('gym.jpg')

img_n = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_n, cmap = 'gray')
```

```python

img_g = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img_g, cmap = 'gray')
```

```python
def gasuss_noise(image, mean=0, var=0.001):
    image = np.array(image/255, dtype=float)
    noise = np.random.normal(mean, var ** 0.5, image.shape)
    out = image + noise
    if out.min() < 0:
        low_clip = -1.
    else:
        low_clip = 0.
    out = np.clip(out, low_clip, 1.0)
    out = np.uint8(out*255)

    return out
```

```python
img_noise = gasuss_noise(img_n)
plt.imshow(img_noise)
```

```python
img_median_blur = cv2.medianBlur(img_noise, 7)
plt.imshow(img_median_blur)
```

```python
img_blur = cv2.blur(img_noise, (13, 13))
plt.imshow(img_blur)
```

```python
ret, tresh = cv2.threshold(img_g, 150, 255, cv2.THRESH_BINARY)
plt.imshow(tresh, cmap = 'gray')
```

```python
img_misha = cv2.cvtColor(cv2.imread('misha.jpg'), cv2.COLOR_BGR2GRAY)
plt.imshow(img_misha, cmap='gray')
```

```python
ret, tresh = cv2.threshold(img_misha, 80, 255, cv2.THRESH_BINARY)
plt.imshow(tresh, cmap='gray')
```

```python
tresh = cv2.adaptiveThreshold(img_misha, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 31, 10)
plt.imshow(tresh, cmap='gray')
```
